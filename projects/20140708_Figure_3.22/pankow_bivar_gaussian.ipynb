```python
random.seed(0)
# Generate data
mu1, mu2, sigma1, sigma2 = 0, 0, 2, 1
g1 = numpy.random.normal(mu1, sigma1, 1e6)
g2 = numpy.random.normal(mu2, sigma2, 1e6)

# Let's look at it in the one dimensional profile
hist(g2, histtype='stepfilled', label='Gaussian$(\mu=%1.2f, \sigma=%1.2f)$' % (mu2, sigma2), alpha=0.5, normed=True, color='b', bins=20)
hist(g1, histtype='stepfilled', label='Gaussian$(\mu=%1.2f, \sigma=%1.2f)$' % (mu1, sigma1), alpha=0.5, normed=True, color='r', bins=20)
xlim([-10, 10])
legend();
```

```python
# Rotate appropriately
alpha = pi/4
gx, gy = (g1-mu1)*cos(alpha)+(g2-mu2)*sin(alpha), -(g2-mu2)*sin(alpha)+(g1-mu2)*cos(alpha)

# Histogram 1 dimensional again
hist(gx, histtype='stepfilled', alpha=0.5, normed=True, color='b', bins=20)
hist(gy, histtype='stepfilled', alpha=0.5, normed=True, color='r', bins=20);
xlim([-8, 8])
```

```python
# ...and for the finale, we do the 2-D histogram
?histogram2d

# Read the documentation, this gives you back a 2-dimensional histogram and the x and y bins used to create it
hgrm, xe, ye = histogram2d(gx, gy, bins=50, normed=True)

# To display this properly, we first "grid" the x and y edges to form a mesh
xgr, ygr = meshgrid(xe, ye)

# Then pcolormesh is used for (possibly uneven) binning. The documentation is explicit about imshow (which could also be used here).
# imshow is really for image data, and so its 'bins' are actually pixels --  by definition this means that bins have to be equal width
# imshow has other peculariities which make it difficult to use for two dimensional data
pcolormesh(xgr, ygr, hgrm)
xlim([-4.5, 4.5])
ylim([-4.5, 4.5])

# Let's also calculate the covariance (what the authors call \sigma_{xy}
covar = cov(gx, gy)

# The matrix returned from this is 2x2, with the trace being \sigma_x^2, \sigma_y^2 and the off diagnonal terms are \sigma_{xy}.
# Recall the author's convention that \sigma_{xy} \propto \sigma_{x,y}^2, e.g. it is a quantity in quadrature despite the lack of
# a power.

# Since the covariance operation is normally symmetric under interchange of the variables, so is the matrix.
print covar[0, 1] == covar[1,0]

# Compare to known values
print covar
print std(gx)**2
print std(gy)**2
```

```python
# The contours are little more tricky. We're actually overlaying the contours ontop of the figure we've already created.
# The ipython cell structure isolates plots within cells (but not variables!) so we'll need to recreate.
# NOTE: I'm adding a colorbar for reference (this isn't part of figure 3.22)
pcolormesh(xgr, ygr, hgrm)
colorbar()
xlim([-4.5, 4.5])
ylim([-4.5, 4.5])

# Rather than draw ellipses at the specified widths, I've elected to demonstrate a water filling algorithm
#
# Then use the contour function, with three specified contours (e.g. where the cumulant of the histogram equals 1, 2, and 3
# sigma on a normal Gaussian.

# The copy is so that we don't affect the plotting done later
vals = sorted(hgrm.copy().flatten(), reverse=True)

cntrs = []
norm_val = sum(vals)
for val in [0.682, 0.954, 0.997]:
    cntrs.append(vals[searchsorted(cumsum(vals)/norm_val, val)])

# Note that the contours here are *also* ellipses with the *same* rotation angle.

# Also note that the gridding size is one larger than the histogram in btoh x and y direction, so we have to chop off the
# end, else contour gets very confused.
contour(xgr[:-1,:-1], ygr[:-1,:-1], hgrm, levels=cntrs, linewidth=5, colors='k');
```
